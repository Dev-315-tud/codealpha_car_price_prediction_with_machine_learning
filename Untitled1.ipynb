MODEL TRAINING

```python
#Machine_learning_library
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
```

```python
x=df.drop(['Selling_Price','Car_Name','Fuel_Type','Selling_type','Transmission'] ,axis=1)
x
```

```python
y=df['Selling_Price']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2 , random_state=42)
```

```python
#Linear Regression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)


```

```python
#Linear Regression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)


```

```python
print("linear Regression,Prediction:",y_pred)

modelp = model.score(x_test, y_test)  
acc = f"Linear  Regression model accuracy : {modelp * 100:.2f}%"
acc
```
